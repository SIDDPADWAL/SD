```python
sns.kdeplot(data=data[data['Survived'] == 0]['Age'], label='Did Not Survive',color='blue')
sns.kdeplot(data=data[data['Survived'] == 1]['Age'], label='Survived',color='orange')
plt.xlabel("Age")
plt.title("Age Distribution by Survival Status")
plt.legend()
plt.show()
```
